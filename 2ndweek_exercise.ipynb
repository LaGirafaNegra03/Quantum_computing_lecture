ライブラリのインポート

```python
import qiskit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
#from qiskit.visualization import visualize_transition

from math import sqrt
from math import pi
```

量子回路の定義

```python
qc = qiskit.QuantumCircuit(5)

state_initial = [1/sqrt(2),-1/sqrt(2)]  #initialize to the correct state
for x in range(5):
    qc.initialize(state_initial,x)
qc.id(0)    
qc.t(1)  #select the correct gate 
qc.s(2)  #select the correct gate
qc.rz(pi/2,3)  #select the correct gate
qc.z(4)  #select the correct gate

state = Statevector.from_instruction(qc)

qc.draw("mpl")
```

ブロッホ球のプロット

```python
plot_bloch_multivector(state, title="qubit0 = initial |-> state, qubit1 = state after T gate, qubit2 = state after S gate, qubit3 = state after Rz(pi/2) gate, qubit3 = state after Z gate")
```
